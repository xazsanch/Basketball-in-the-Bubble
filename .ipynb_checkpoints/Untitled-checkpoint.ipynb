```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
from datetime import datetime, time
df = pd.read_csv('2018-2019 Game Log Playoffs.csv')
df.head()
df = df.drop(df.columns[[0]], axis = 1)
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
pd.set_option('display.max_columns', None)
df
```

```python
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (7, 7),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
```

```python
df2 = pd.read_csv('2019-2020 Game Log Playoffs.csv')
df2 = df.drop(df.columns[[0]], axis = 1)
cols = df2.columns.tolist()
cols = cols[-1:] + cols[:-1]
df2 = df[cols]
pd.set_option('display.max_columns', None)
df2
```

```python
labels = ['']
x=np.arange(len(labels))
width=0.35

pts1 = round(np.mean(df['PTS']),2)
pts2 = round(np.mean(df2['PTS']),2)

fig, ax = plt.subplots(figsize=(7,7))

rects1 = ax.bar(x-width/2,pts1, width, color='blue',label='2019')
rects2 = ax.bar(x+width/2,pts2, width, color='orange',label='2020')

ax.set_ylabel('Average Points')
ax.set_title('Average Points in the Post-Season')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)



fig.tight_layout()
plt.show()
```
